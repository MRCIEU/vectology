```python
df_raw = pd.read_csv(df_path)
df_raw.info()
df_raw
```

```python
UNMAPPED_MODELS = ["BLUEBERT-EFO", "BlueBERT", "BioSentVec"]
df = df_raw[df_raw["model_name"].isin(UNMAPPED_MODELS)]
df
```

```python
tbl_sum = (
    df[["model_name", "MAPPING_TYPE", "group_rank"]]
    .groupby(["MAPPING_TYPE", "model_name"])
    .describe()
)
tbl_sum
```

```python
tbl_sum.to_csv("~/Downloads/describe_top100.csv")
```
